```python
import xml.etree.ElementTree as ET
from pathlib import Path
from collections import defaultdict

input_folder = Path(r"C:\Users\culle\Downloads\AI-574-Group4-Main\data\ami_public_manual_1.6.2\words")

output_folder = Path(r"C:\Users\culle\Downloads\AI-574-Group4-Main\Transcripts")

def parse_xml_to_words(xml_path):
    tree = ET.parse(xml_path)
    root = tree.getroot()

    words = []
    for w in root.iter("w"):
        if w.text:
            start = float(w.attrib.get("starttime", 0))
            words.append((start, w.text))

    return words

meetings = defaultdict(list)

for xml_file in input_folder.glob("*.words.xml"):
    meeting_id = xml_file.name.split(".")[0]
    words = parse_xml_to_words(xml_file)
    meetings[meeting_id].extend(words)

for meeting_id, word_list in meetings.items():
    word_list.sort(key = lambda x: x[0])
    text = " ".join(word for _, word in word_list)

    output_path = output_folder / f"Tran{meeting_id}.Mix-Headset.txt"

    with open(output_path, "w", encoding = "utf-8") as f:
        f.write(text)

        print(f"Saved: {output_path}")
```
